```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))

plt.title("Grades based on study time")
sns.regplot(data = df, x="studytime", y="G1",  color="orange",label="G1: First Period Grades", 
            scatter= False, ci= None)
sns.regplot(data = df, x="studytime", y="G2", color="red", label="G2: Second Period Grades",
           scatter= False, ci= None)
sns.regplot(data = df, x="studytime", y="G3", color="blue", label="G3: Final Period Grades",
            scatter= False, ci= None)
plt.xlabel("Study Time")
plt.ylabel("Grades")

plt.legend(loc="upper left")
plt.show()
```

```python
#  Box plot of on student's age
plt.figure(figsize=(8, 6))
sns.boxplot(y=students_df['age'], color='blue')
plt.title('Box Plot of Age')
plt.xlabel("Student's age")
plt.ylabel("")
plt.show()
```

```python

students_df['address'] = students_df['address'].map({'U':'Urban', 'R':'Rural'})

def plot_horizontal_bar_chart(df, column):
    count_series = df[column].value_counts()
    plt.figure(figsize=(6, 4))
    barplot = sns.barplot(x=count_series.values, y=count_series.index, orient='h', palette='viridis')
      
    for i, v in enumerate(count_series.values):
        barplot.text(v + 0.1, i, str(v), color='black', va='center')
       
    barplot.set(ylabel=None)    
    plt.xlabel('Count')
    plt.title(f'Distribution of {column.capitalize()}', fontsize=10)
    plt.show()
plot_horizontal_bar_chart(students_df, 'address')
```

```python
students_df['sex'] = students_df['sex'].map({'M':'Male', 'F':'Female'})

def plot_horizontal_bar_chart(df, column):
    count_series = df[column].value_counts()
    plt.figure(figsize=(6, 4))
    barplot = sns.barplot(x=count_series.values, y=count_series.index, orient='h', palette='viridis')
      
    for i, v in enumerate(count_series.values):
        barplot.text(v + 0.1, i, str(v), color='black', va='center')
       
    barplot.set(ylabel=None)    
    plt.xlabel('Count')
    plt.title(f'Distribution of {column.capitalize()}', fontsize=10)
    plt.show()
plot_horizontal_bar_chart(students_df, 'sex')
```

```python
import pandas as pd

# Assuming 'df' is your DataFrame
df_melted = pd.melt(df, id_vars=['G3'], value_vars=['Mjob', 'Fjob'], var_name='Job', value_name='Job Type')

# Create the catplot
plot = sns.catplot(data=df_melted, x='G3', y='Job Type', hue='Job', kind='box')

# Set the axis labels and title
plot.set(xlabel='Score', ylabel='Job Type')
plot.fig.suptitle('Distribution of Mjob and Fjob', fontsize=10)

# Adjust layout to make room for the title
plot.fig.tight_layout(rect=[0, 0, 1, 0.95])

# Adjust the legend position to the right corner
plot._legend.set_bbox_to_anchor((1, 1))
plot._legend.set_frame_on(True)  # Optional: add a frame around the legend

# Show the plot
sns.show()
```

```python
education_mapping = {0: 'None', 1: 'Primary education', 2: 'Middle school', 3: 'Secondary education', 4: 'Higher education'}

# Create new columns with the mapped values
df['Medu_desc'] = df['Medu'].map(education_mapping)
df['Fedu_desc'] = df['Fedu'].map(education_mapping)

# Melt the DataFrame
df_melted = pd.melt(df, id_vars=['G3'], value_vars=['Medu_desc', 'Fedu_desc'], var_name='Education', value_name='Education Level')

# Create the catplot
plot = sns.catplot(data=df_melted, x='G3', y='Education Level', hue='Education', kind='box')

# Set the axis labels and title
plot.set(xlabel='Score', ylabel='Education Level')
plot.fig.suptitle('Distribution of Medu and Fedu', fontsize=10)


plot.fig.tight_layout(rect=[0, 0, 1, 0.95])

# Customize box colors
for ax in plot.axes.flat:
    for patch in ax.artists:
        hue = patch.get_facecolor()
        if 'Medu' in str(patch):
            if patch.get_ydata()[0] == 'None':  # Assuming 'None' is at home category
                patch.set_facecolor('Greys_d')
            else:
                patch.set_facecolor('Greys')
        elif 'Fedu' in str(patch):
            if patch.get_ydata()[0] == 'None':  # Assuming 'None' is at home category
                patch.set_facecolor('Blues_d')
            else:
                patch.set_facecolor('light_b')

# Adjust the legend position to the right corner
plot._legend.set_bbox_to_anchor((1, 1))
plot._legend.set_frame_on(True)  # Optional: add a frame around the legend

# Show the plot
plt.show()
```

```python
columns_to_select_1 = ["G3", "paid", "activities", "nursery", "higher", "internet", "schoolsup"]
df_1= df[columns_to_select_1]
display(df_1)
```

```python
df_2=df_1.copy()

# Encode categorical variables
df_2['paid'] = df_2['paid'].map({'yes': 1, 'no': 0})
df_2['activities'] = df_2['activities'].map({'yes': 1, 'no': 0})
df_2['nursery'] = df_2['nursery'].map({'yes': 1, 'no': 0})
df_2['higher'] = df_2['higher'].map({'yes': 1, 'no': 0})
df_2['internet'] = df_2['internet'].map({'yes': 1, 'no': 0})
df_2['schoolsup'] = df_2['schoolsup'].map({'yes': 1, 'no': 0})

# Calculate correlation
correlation_matrix = df_2.corr()

# Print the correlation of G3 with other variables
print(correlation_matrix['G3'])
```

```python
# Print the correlation of G3 with other variables
print(correlation_matrix['G3'])# Generate a correlation matrix plot
plt.figure(figsize=(6, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation matrix of school practices on Students Final exam grades')
plt.show()
```
